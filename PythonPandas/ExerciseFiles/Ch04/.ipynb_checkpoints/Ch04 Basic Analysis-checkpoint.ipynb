# Basic Analysis

```python
import pandas as pd
```

```python
oo = pd.read_csv('../data/olympics.csv', skiprows=4)
oo.head()
```

### value_counts()

```python
oo.Edition.value_counts()
```

```python
oo.Gender.value_counts(ascending=True,dropna=False)
```

```python
oo[["Edition", "Gender"]].value_counts(ascending = True)
```

### sort_values()

```python
ath = oo.Athlete.sort_values()
ath
```

```python
oo.sort_values(by=['Edition','Athlete'])
```

### Boolean indexing

```python
oo.head()
```

```python
oo[(oo.Medal == 'Gold') & (oo.Gender == 'Women')]
```

```python
x = oo[(oo.Medal == 'Gold') & (oo.Gender == 'Women')]
x.groupby('Edition')['Athlete'].nunique().reset_index().sort_values('Athlete').sort_values('Athlete', ascending = False)
```

### String handling

```python
oo[oo.Athlete.str.contains('Florence')]
```

```python
oo[oo.Athlete.str.contains('Florence')]
```

```python
oo[oo.Athlete.str.contains('OWENS, Jesse')].Event.value_counts()
```

```python
a = oo[(oo['Gender'] == "Men") & (oo['Medal'] == "Gold") & (oo['Discipline'] == 'Badminton') & (oo['Event'] == 'singles')][['NOC', 'Athlete']].value_counts().reset_index()#.sort_values(by = ['Athlete'])
a.sort_values(by = ['Athlete'])
```

```python
b = oo[(oo.Edition >= 1984) & (oo.Edition <= 2008)]
b = b.value_counts('NOC')#.sort_values(by = ['NOC'], ascending = False)#groupby('City')['Medal'].nunique()
b.head(3)

```

```python
c = oo[(oo.Gender == 'Men') & (oo.Medal == 'Gold') & (oo.Discipline == 'Athletics') & ((oo.Event == '100m') | (oo.Event == 'Sprint indivual'))]
c.sort_values('Edition', ascending = False)[['City', 'Edition', 'Athlete', 'NOC']]


```
